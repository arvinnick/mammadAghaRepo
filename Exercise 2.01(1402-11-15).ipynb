```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv(r'D:\Course statistics\Exercise 2/BX-Book-Ratings1.csv',encoding = "ISO-8859-1")
print(data)
```

```python
data['ISBN'] = data['ISBN'].astype('str')
groupID = data.groupby("User-ID")["ISBN"].apply(lambda x: x.tolist()).reset_index()
groupID = groupID.rename(columns={"Transaction":"Products"})
# print(groupID)
dataset = list(groupID['ISBN'])
dataset = dataset[0:20000]
print(dataset)
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(dataset).transform(dataset)
df = pd.DataFrame(a_data,columns=a.columns_)
df = df.replace(False,0)
print(df)
```

```python
#set a threshold value for the support value and calculate the support value.
df = apriori(df, min_support = 0.2, use_colnames = True, verbose = 1)
print(df)
```

```python
#Let's view our interpretation values using the Associan rule function.
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.6)
print(df_ar)
```
