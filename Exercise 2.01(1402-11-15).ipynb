```python
######################################################### 20 rule with highest lift
from mlxtend.frequent_patterns import association_rules

association_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
association_rules = association_rules.sort_values(by = 'lift',ascending= False)
association_rules.head(20)
```
